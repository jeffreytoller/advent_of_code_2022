# Day 2

Read file and split into list

```python
input = open('Day02_Data.txt')
input = input.read().split('\n\n')
list_input = []
for i,v in enumerate(input):
    list_input.append(v.split("\n"))
```

Read list into pandas dataframe and set NaNs to 0 to allow summing of rows

```python
import pandas as pd
df = pd.DataFrame(list_input[0])
df = pd.DataFrame(list(df[0].str.split(' ')),columns=['opponent_move','player_move'])
df
```

```python
import itertools
opponent_moves = ['A','B','C']
player_moves = ['X','Y','Z']
wl_df = pd.DataFrame(list(itertools.product(opponent_moves, player_moves)),columns=['opponent_move','player_move'])
wl_df
```

```python
wl_df['win_points'] = [3,6,0,0,3,6,6,0,3]
wl_df
```

```python
pp_df = pd.DataFrame([['X',1],['Y',2],['Z',3]],columns=['player_move','play_points'])
pp_df
```

```python
df = df.merge(how='left',on=['opponent_move','player_move'],right=wl_df)
```

```python
df = df.merge(how='left',on='player_move',right=pp_df)
```

```python
df['total_points'] = df.win_points+df.play_points
```

```python
df
```

```python
df.total_points.sum()
```

```python
st_df = pd.DataFrame([['X',0],['Y',3],['Z',6]],columns=['player_move','actual_win_points'])
```

```python
df = df.merge(how='left',on='player_move',right=st_df)
```

```python
df
```

```python
df = df.merge(how='left',right=wl_df,left_on=['opponent_move','actual_win_points'],right_on=['opponent_move','win_points'])
df = df.merge(how='left',left_on='player_move_y',right_on = 'player_move',right=pp_df)
df['actual_total_points'] = df.play_points_y + df.win_points_y
df
```
